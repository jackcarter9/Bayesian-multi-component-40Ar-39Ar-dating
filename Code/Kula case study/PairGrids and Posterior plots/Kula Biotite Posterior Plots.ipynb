# Kula 3

```python
df1 = pd.read_csv('Kula_biotite_3_posterior.csv')
```

```python
df1.head()
```

```python
age_model = df1['age_model'].values
age_low = df1['age_model_low'].values
age_high = df1['age_model_high'].values

f_model = df1['frel_model'].values
f_low = df1['frel_model_low'].values
f_high = df1['frel_model_high'].values

age_data = df1['age_data'].values 
age_err = df1['age_err'].values
frel_data = df1['frel_data'].values


mod1 = np.insert(age_model, 0, age_model[0])
mod1_low = np.insert(age_low, 0, age_low[0])
mod1_high = np.insert(age_high, 0, age_high[0])

fmod1 = f_model
f_low1 = f_low
f_high1 = f_high

ymod1 = np.insert(fmod1,0, 0)

age1 = age_data
err1 = age_err
frel1 = frel_data

y1 = np.insert(frel1, 0, 0)
age_plot1 = np.insert(age1, 0, age1[0])
age_err_plot1 = np.insert(err1, 0, err1[0])
```

```python
plt.style.use('classic')
plt.figure(figsize = (5,5))
plt.fill_between(y1, age_plot1 - age_err_plot1, 
               age_plot1 + age_err_plot1, 
                step = 'pre', interpolate = True, 
                color = 'grey', alpha = 0.7, 
                label = 'Data')

plt.fill_between(ymod1, mod1_low, mod1_high, 
                step = 'pre', interpolate = True, 
                color = 'red', alpha = 0.7, 
                label = '95% Credible\nInterval')

plt.xlabel('Cumulative $^{39}$Ar released')
plt.ylabel('Apparent Age (Ma)')
plt.ylim(65, 130)
```
