```python
import numpy as np 
import matplotlib.pyplot as plt
import seaborn as sns 
plt.style.use("classic")
```

```python
import os
import pandas as pd
```

```python
cwd = os.chdir("/Users/jackcarter/Documents/Data_Mixtures")

df1 = pd.read_csv("Kula_mix1.csv")
df2 = pd.read_csv("Kula_mix2.csv")
df1.columns = ["Age$_1$", "Age$_2$", "$\chi_1$", "$\chi_2$"]
df2.columns = ["Age$_1$", "Age$_2$", "$\chi_1$", "$\chi_2$"]
```

```python
plt.figure(figsize = (20,20))
g = sns.PairGrid(df1[9000:-800])
g = g.map_diag(plt.hist, facecolor = 'mediumaquamarine', edgecolor = 'k', bins = 15)
g = g.map_offdiag(plt.scatter,facecolor = 'mediumaquamarine', edgecolor = 'k')
#g.axes[,0].set_ylim(-10,10)
g.axes[0,0].set_xlim(0,200)
g.axes[0,1].set_xlim(0,200)
g.axes[0,2].set_xlim(0,1)
g.axes[0,3].set_xlim(0,1)

g1 = sns.PairGrid(df2[18000:-1600])
g1 = g1.map_diag(plt.hist, facecolor = 'yellow', edgecolor = 'k', bins = 15)
g1 = g1.map_offdiag(plt.scatter,facecolor = 'yellow', edgecolor = 'k')
g1.axes[0,0].set_xlim(0,200)
g1.axes[0,1].set_xlim(0,200)
g1.axes[0,2].set_xlim(0,1)
g1.axes[0,3].set_xlim(0,1)
#g.axes.xticks(np.arange(min(x), max(x)+1, 1.0))
plt.savefig('Pairplot_12.pdf')
```
