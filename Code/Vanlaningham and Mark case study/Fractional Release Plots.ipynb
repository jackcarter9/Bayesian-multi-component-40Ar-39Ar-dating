```python
# ACS + TCR
df4 = pd.read_csv("tcr_acs_fr.csv", engine = "python")
```

```python
fig, ax = plt.subplots(2,2, figsize = (7.48031, 7.48031))




# ACS + TCR
ax[0,0].plot(df4["temp"].values, df4["data"].values, 'ok-', 
            label = 'Data')
ax[0,0].fill_between(df4["temp"].values, df4["Model"].values + df4["m_err"].values, 
                                                                  df4["Model"].values - 
                    df4["m_err"].values, alpha = 0.4, color = 'grey', 
                    label = '95%\nCredible\nInterval')

ax[0,0].plot(df4["temp"].values, df4["Model"].values, ls = '--', 
             color = 'grey',
            label = 'Mean\nposterior\nmodel')
ax[0,0].legend(loc = 'best', 
              fontsize = 'x-small')


# ACS + HDB!1
ax[0,1].plot(df1["temp"].values, df1["data"].values, 'ok-', 
            label = 'Data')
ax[0,1].fill_between(df1["temp"].values, df1["Model"].values + df1["m_err"].values, 
                                                                  df1["Model"].values - 
                    df1["m_err"].values, alpha = 0.4, color = 'grey', 
                    label = '95% Credible Interval')

ax[0,1].plot(df1["temp"].values, df1["Model"].values, ls = '--', 
             color = 'grey',
            label = 'Mean posterior model')



# HDB1 + TCR
ax[1,0].plot(df3["temp"].values, df3["data"].values, 'ok-', 
            label = 'Data')
ax[1,0].fill_between(df3["temp"].values, df3["Model"].values + df3["m_err"].values, 
                                                                  df3["Model"].values - 
                    df3["m_err"].values, alpha = 0.4, color = 'grey', 
                    label = '95% Credible Interval')

ax[1,0].plot(df3["temp"].values, df3["Model"].values, ls = '--', 
             color = 'grey',
            label = 'Mean posterior model')



# HDB1 + TCR + ACS
ax[1,1].plot(df2["temp"].values, df2["data"].values, 'ok-', 
            label = 'Data')
ax[1,1].fill_between(df2["temp"].values, df2["Model"].values + df2["m_err"].values, 
                                                                  df2["Model"].values - 
                    df2["m_err"].values, alpha = 0.4, color = 'grey', 
                    label = '95% Credible Interval')

ax[1,1].plot(df2["temp"].values, df2["Model"].values, ls = '--', 
             color = 'grey',
            label = 'Mean posterior model')

# Set axis labels 
#ax[0,0].set_xlabel('Temperature ($^{o}C$)')
ax[1,0].set_xlabel('Temperature ($^{o}C$)')
#ax[0,1].set_xlabel('Temperature ($^{o}C$)')
ax[1,1].set_xlabel('Temperature ($^{o}C$)')

ax[1,0].set_ylabel('Fractional Release $^{39}$Ar')
ax[0,0].set_ylabel('Fractional Release $^{39}$Ar')
plt.tight_layout()
ax[0,0].set_ylim(0.0, 1.05)
ax[1,0].set_ylim(0.0, 1.05)
ax[1,1].set_ylim(0.0, 1.05)
ax[0,1].set_ylim(0.0, 1.05)
plt.savefig('frel_vm.pdf')
```
